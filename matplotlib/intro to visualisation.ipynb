```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = x + x
z = x + y
```

```python
plt.plot(x,y, z)
```

```python
x = np.linspace(0,10, 20)
y = x  * x
z = x + y
```

```python
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("x and y axis graph")
```

```python
plt.plot(x,y,z)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("x,y and z graph")
plt.legend(['x','y'])
```
