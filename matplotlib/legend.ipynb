```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10,20)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x,y,x,z)
plt.legend(['sine', 'cosine'])
```

```python
plt.plot(x,y,x,z)
plt.legend(['sine', 'cosine'], loc = 1)
```

```python
plt.plot(x,y,y,z)
plt.legend(['sine','cosine'], loc = 10) #loc ranges from 1 - 10
```
