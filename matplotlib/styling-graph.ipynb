```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = np.cos(x)
```

```python
plt.title("Cosine graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y, 'g', linewidth=3)
```

```python
plt.title("Cosine graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y, 'g', linewidth=3, alpha = 0.5)
```

```python
plt.title("Cosine graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y, 'g', linewidth=3, alpha = 0.5, linestyle = "-.")
```

```python
plt.title("Cosine graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y, 'g', linewidth=3, alpha = 0.5, marker = "o", markersize = 10)
```

```python
plt.title("Cosine graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y, 'g', linewidth=3, alpha = 0.5, marker = "1", markersize = 20)
```

```python
plt.title("Cosine graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,  linewidth=3,  marker = "o", markersize = 20, markerfacecolor = 'red', markeredgecolor = "green")
```

```python
plt.title("Cosine graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,  linewidth=3,  marker = "o", markersize = 20, markerfacecolor = 'red', markeredgecolor = "green", markeredgewidth = 3)
```
