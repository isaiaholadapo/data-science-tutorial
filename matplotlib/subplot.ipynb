```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,40, 10)
y = x * 2
z = y ** 2
```

```python
plt.subplot(3,2,1)
plt.plot(x,y)
plt.title(1)
plt.subplot(3,2,2)
plt.plot(x,z)
plt.title(2)
plt.subplot(3,2,3)
plt.plot(y,z)
plt.title(3)
plt.subplot(3,2,4)
plt.plot(x,y,z)
plt.title(4)
plt.tight_layout()
```
